```python
from Bio import SeqIO

def extract_sequence_by_name(fasta_file, sequence_name):
    for record in SeqIO.parse(fasta_file, "fasta"):
        if record.id == sequence_name:
            return record.seq
    return None

# Example usage
fasta_file = "/data/jenuwein/group/erikson/20220105_Nanopore/6_Assembly4.0/MSR_regions.fasta"
sequence_name = "contig_1071"
sequence = extract_sequence_by_name(fasta_file, sequence_name)
if sequence:
    print(f"Sequence found: {sequence}")
else:
    print("Sequence not found.")
```

```python
from Bio import SeqIO

def extract_sequence_by_name(fasta_file, sequence_name):
    for record in SeqIO.parse(fasta_file, "fasta"):
        if record.id == sequence_name:
            return record.seq
    return None

def find_subsequence_location(sequence, subsequence):
    return sequence.find(subsequence)

# Example usage
fasta_file = "/data/jenuwein/group/erikson/20220105_Nanopore/6_Assembly4.0/MSR_regions.fasta"
sequence_name = "contig_1071"
subsequence = "TATAAAAG"

sequence = extract_sequence_by_name(fasta_file, sequence_name)
if sequence:
    location = find_subsequence_location(sequence, subsequence)
    if location != -1:
        print(f"Subsequence '{subsequence}' found at position: {location}")
    else:
        print(f"Subsequence '{subsequence}' not found.")
else:
    print("Sequence not found.")
```

```python
from Bio import SeqIO

def extract_sequence_by_name_and_coordinates(fasta_file, sequence_name, start, end):
    for record in SeqIO.parse(fasta_file, "fasta"):
        if record.id == sequence_name:
            return record.seq[start:end]
    return None

# Example usage
fasta_file = "/data/jenuwein/group/erikson/20220105_Nanopore/6_Assembly4.0/MSR_regions.fasta"
sequence_name = "contig_1071"
start = 28000  # Replace with your start coordinate
end = 29000    # Replace with your end coordinate

sequence = extract_sequence_by_name_and_coordinates(fasta_file, sequence_name, start, end)
if sequence:
    print(f"Sequence found: {sequence}")
else:
    print("Sequence not found.")
```
